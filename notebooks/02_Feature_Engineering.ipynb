```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r"C:\Projects\food-delivery-regression\data\raw\Food_Delivery_Time_Prediction.csv")
df.head()
```

```python
df.drop(columns = ['Order_ID'], inplace= True)
df.columns
```

```python
feature_cols = [
    'Distance',
    'Delivery_Person_Experience',
    'Restaurant_Rating',
    'Customer_Rating',
    'Order_Cost',
    'Tip_Amount',
    'Traffic_Conditions',
    'Vehicle_Type',
    'Weather_Conditions',
    'Order_Time',
    'Order_Priority'
]

X = df[feature_cols]
y = df['Delivery_Time']
```

```python
X_encoded = pd.get_dummies(
    X,
    columns=[
        'Traffic_Conditions',
        'Vehicle_Type',
        'Weather_Conditions',
        'Order_Time',
        'Order_Priority'
    ],
    drop_first=True
)
X_encoded
```

```python
#   Traffic has an ordinal nature, but its impact on delivery time is non-linear and asymmetric, so we encode it categorically to avoid imposing 
#   false linear structure.
```

```python
#   No ordinal encoding, because ordinal encoding would incorrectly enforce equal linear spacing between traffic levels, which linear regression 
#   cannot relax.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
num_cols = [
    'Distance',
    'Delivery_Person_Experience',
    'Restaurant_Rating',
    'Customer_Rating',
    'Order_Cost',
    'Tip_Amount'
]
```

```python
scaler = StandardScaler()
```

```python
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```
